Applying Pipeline + Cross Validation

```python
pipeline_lr = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LogisticRegression(max_iter=1000))
])

scores_lr = cross_val_score(
    pipeline_lr,
    X,
    Y,
    cv=5
)

print(scores_lr)
print(scores_lr.mean())
```

```python
pipeline_svm = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', SVC(kernel='linear'))
])

scores_svm = cross_val_score(
    pipeline_svm,
    X,
    Y,
    cv=5
)

print(scores_svm)
print(scores_svm.mean())
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

pipeline_rf = Pipeline([
    ('rf', RandomForestClassifier(
        n_estimators=100,
        random_state=42
    ))
])

scores_rf = cross_val_score(
    pipeline_rf,
    X,
    Y,
    cv=cv
)

print("RF CV Scores:", scores_rf)
print("Mean Accuracy:", scores_rf.mean())
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

pipeline_lr = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LogisticRegression(max_iter=1000))
])

scores_lr = cross_val_score(
    pipeline_lr,
    X,
    Y,
    cv=cv
)

print("LR CV Scores:", scores_lr)
print("Mean Accuracy:", scores_lr.mean())
```
